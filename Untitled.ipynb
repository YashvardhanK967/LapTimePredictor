```python
import fastf1
import pandas as pd
import numpy as np
import os
import time
import socket  # <-- NEW: Allows us to control the network connection

# THE KILL SWITCH: Force any frozen connection to crash after 30 seconds
socket.setdefaulttimeout(30.0)

# 1. Setup cache
os.makedirs('f1_cache', exist_ok=True)
fastf1.Cache.enable_cache('f1_cache')

TARGET_YEAR = 2025
print(f"Fetching schedule for the {TARGET_YEAR} season...")
schedule = fastf1.get_event_schedule(TARGET_YEAR)

all_races = []

for index, event in schedule.iterrows():
    if event['EventFormat'] == 'testing':
        continue
        
    event_name = event['EventName']
    print(f"\n--- Processing: {event_name} ---")
    
    max_retries = 3
    race_success = False
    
    # NEW: Self-healing retry loop
    for attempt in range(max_retries):
        try:
            print(f"  -> Attempt {attempt + 1}: Connecting and downloading...")
            session = fastf1.get_session(TARGET_YEAR, event_name, 'R')
            session.load(telemetry=False, weather=True) 
            
            print("  -> Extracting and merging data...")
            laps = session.laps
            weather = laps.get_weather_data()
            
            laps_clean = laps.reset_index(drop=True)
            weather_clean = weather.reset_index(drop=True)
            dataset = pd.concat([laps_clean, weather_clean], axis=1)
            
            dataset['is_pit_lap'] = dataset['PitInTime'].notnull().astype(int)
            
            features = ['Driver', 'LapNumber', 'Stint', 'Compound', 
                        'TrackTemp', 'AirTemp', 'Rainfall', 'is_pit_lap']
            race_data = dataset[features].copy()
            
            race_data['Track'] = event_name
            race_data['TireAge'] = race_data.groupby(['Driver', 'Stint']).cumcount() + 1
            race_data = race_data.drop(columns=['Stint'])
            
            all_races.append(race_data)
            print("  -> ✅ Race successfully processed!")
            race_success = True
            
            # Sleep to prevent rate-limiting on the next race
            time.sleep(5) 
            break  # Break out of the retry loop because it worked!
            
        except Exception as e:
            print(f"  -> ⚠️ Attempt {attempt + 1} failed (Timeout/Error: {e})")
            time.sleep(5) # Wait a few seconds before trying again
            
    if not race_success:
        print(f"  -> ❌ ERROR: Completely skipping {event_name} after {max_retries} failed attempts.")

# ---------------------------------------------------------
# MASTER CLEANING & EXPORT
# ---------------------------------------------------------
print("\nCombining and cleaning all season data...")
final_data = pd.concat(all_races, ignore_index=True)

final_data = final_data.replace(r'^\s*$', np.nan, regex=True)
final_data = final_data.replace(['nan', 'NaN', 'None', 'null'], np.nan)
final_data = final_data.dropna()

final_data = final_data[['Track', 'Driver', 'LapNumber', 'Compound', 'TireAge', 
                         'TrackTemp', 'AirTemp', 'Rainfall', 'is_pit_lap']]

csv_filename = f'f1_pitstops_{TARGET_YEAR}_full_season.csv'
final_data.to_csv(csv_filename, index=False)

print(f"\n✅ Success! Full season CSV saved to:\n{os.path.abspath(csv_filename)}\n")
```
